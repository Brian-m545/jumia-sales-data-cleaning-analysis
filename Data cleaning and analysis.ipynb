```python

#checking outliers for current prices
sns.boxplot(y='Current prices', data=df)
```

```python
#lower and upper quartile
Q1 = df['Current prices'].quantile(0.25)
Q3 = df['Current prices'].quantile(0.75)
IQR = Q3 - Q1
```

```python
IQR = Q3 - Q1
```

```python
# Interquartile range
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
#removing outliers
outliers = df[(df['Current prices'] < lower_bound) | (df['Current prices'] > upper_bound)]
print("Outliers using IQR method:\n", outliers)
```
